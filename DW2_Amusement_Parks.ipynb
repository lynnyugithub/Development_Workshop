```python
import pandas as pd 
import numpy as np
```

```python
tx_injuries_df  = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/'\
                              'tidytuesday/master/data/2019/2019-09-10/tx_injuries.csv')

safer_parks_df = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/'\
                             'tidytuesday/master/data/2019/2019-09-10/saferparks.csv', encoding = 'cp1252')
```

1) Create new columns using the data from tx_injuries.csv
   +  The first column will be called ticket_purchased and it should identify the individual as a child or an adult  (if age< 18, a child and adult if age>18)
   +  The second column will be called accident_year and by extracting  the year that the injury occurred(injury_date) 
          + remove occurrences that don't have a year
   +  Find the number of instances for each combination of  accident_year and ticket_purchased (including nulls from not being able to parse dates)
   + Create a column "city_state" as a string by combining columns city and st with '_' 
           + For instance city = Austin and state = TX becomes Austin_TX

```python
tx_injuries_df['age'] = tx_injuries_df['age'].str.extract('(\d+)').dropna().astype('int64')

tx_injuries_df.dropna(subset = ['age'], how='all', inplace=True)

tx_injuries_df['ticket_purchased'] = np.where(tx_injuries_df.age < 18, 'child', 'adult')
```

```python
tx_injuries_df['injury_date'] = pd.to_datetime(tx_injuries_df['injury_date'], errors='coerce')

tx_injuries_df['accident_year'] = tx_injuries_df['injury_date'].dt.year.astype('Int64')
```

```python
tx_injuries_df.groupby(['accident_year', 'ticket_purchased']).size()
```

```python
tx_injuries_df['city_state'] = tx_injuries_df['city'] + '_' + tx_injuries_df['st']
```
