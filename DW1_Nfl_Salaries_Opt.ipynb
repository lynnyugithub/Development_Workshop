1) Initial Data Setup

  +  Clean the column names of all columns by converting them to all lower case 

```python
import pandas as pd

salaries_df = pd.read_excel('nfl_salaries.xlsx')

salaries_df.columns = salaries_df.columns.str.lower()

salaries_df.head()
```

2) What is the average salary for quarterbacks (Columnname : Quarterback) by year?

```python
salaries_df.groupby('year')['quarterback'].mean().to_frame('avg_quarterback').reset_index().round({'avg_quarterback': 1})
```

3) What is the average salary by year for each position, excludes special teamers (Columnname : Special Teamer)?
  +  Hint: Transpose the dataset into "long" format to help with this step.

```python
dropped_df = salaries_df.drop('special teamer', axis=1)

dropped_df.head()
```

```python
columns_list = list(salaries_df.columns)
columns_list.remove('year')
columns_list
```

```python
melt_df = pd.melt(salaries_df, id_vars=['year'], value_vars=columns_list,
        var_name='position', value_name='salaries')

melt_df.head()
```

```python
avg_salaries_df = melt_df.groupby(['year', 'position'])['salaries'].mean().to_frame('avg_salaries').reset_index().round(1)

avg_salaries_df.head()
```

4) Convert the average salary by year for each position to a string prefixed with $ (e.g. 100000 becomes $100000)

```python
avg_salaries_df['avg_salaries'] = '$' + avg_salaries_df['avg_salaries'].astype(str)

avg_salaries_df.head()
```
